```python
# Vectorization of Statement (문장의 vector 화)

import pandas as pd
sentens = ['I love my dog.',
             'I love my cat.',
             'I love my dog and love my cat',
             'You love my dog!',
             'Do you think my dog is amazing?']
```

### Bag of Word (BOW)

CountVectorizer는 Python의 scikit-learn 라이브러리에 포함된 클래스로, 텍스트 데이터의 토큰화(tokenization)와 단어 빈도 수를 기반으로 하는 피처 벡터 (feature vector)를 생성하는 데 사용됩니다. 이 클래스는 자연어 처러(Natural Language Processing, NLP)와 텍스트 마이닝에서 널리사용도니다.

주요 기능
- 트큰화 (tokenization) : 문장이나 문서를 개별 단어나 표현으로 분할합니다.
- 단어 빈도 계산 (Word Frequency Counting) : 각 단어가 문서 내에서 나타나는 빈도를 계산합니다.
- 피처 벡터 생성 (Feature Vector Creation) : 각 문서를 단어의 빈도를 나타내는 벡터로 변환합니다.이 벡터는 머신러닝 알고리즘에 입력으로 사용될 수 있습니다.
- 사전 구축 (Vocabulary Building) : 모든 문서에서 사용된 모든 단어의 사전을 만듭니다.

CountVectorizer를 사용하면 텍스트 데이터를 수치적으로 분석할 수 있으며, 이는 감정분석, 주제 모델링, 문서 분류와 같은 다양한 NLP 응용 프로그램에서 중요한 단계입니다. 예를 들어, 스팸 메일 분류, 문서 군집화, 텍스트 기반 추천 시스템 등에 사용된다.

CountVectorizer 주요 파라미터
- min_df : vocabulary에 포함할 최소 발생 빈도
- ngram_range : (1,1) - unigram only, (1,2) - unigram + bigram
- max_features : top max_features 만으로 vocabulary 구성
- token_pattern = (?u)\b\w\w+\b : unocode 영수자 2 글자 이상만 포함




```python
# Text vs Token Matrix 생성
from sklearn.feature_extraction.text import CountVectorizer

# CountVectorizer 객체 생성
count_vectorizer = CountVectorizer()

# sentences 데이터에 대한 피처 변환 수행
features = count_vectorizer.fit_transform(sentens)
features
```

```python
print(f'document 수 : {features.shape[0]}')
print(f'단어수 : {features.shape[1] - 1}')
```

```python
vectorized_sentenses = features.toarray()
vectorized_sentenses
```

```python
# features의 단어 list
# CountVectorizer를 통해 추출한 피처(단어) 이름들을 가져옴
feature_names = count_vectorizer.get_feature_names_out()
feature_names
```

```python
df = pd.DataFrame(vectorized_sentenses, columns=feature_names)
df

# 데이터프레임의 인덱스 이름 지정
df.index.name = 'sentence'
df
```

```python
# TF-IDF (Term frequency - Inverse Document Frequency)
# TF-IDF는 단어의 빈도와 그 단어가 드물게 나타나는 문서에 더 높은 가중치를 부여하는 방식으로 작동한다.
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()

# sentences 데이터에 대한 TF-IDF 기반 치퍼 변환 수행
tfidf_sentences = tfidf_vectorizer.fit_transform(sentens)
tfidf_sentences
```

### Text vs tf-idf matrix 생성

```python
# Text vs tf-idf Matrix 생성
tfidf_vect_sentences = tfidf_sentences.toarray()
tfidf_vect_sentences
```

```python
# TfidVectorizer를 통해 추출한 피처(단어) 이름들을 가져옴
tfidf_feature_names = tfidf_vectorizer.get_feature_names_out()
tfidf_feature_names
```

```python
# TF-IDF 백테화된 문장과 피처 이름을 이용해 DataFrame 생성
df = pd.DataFrame(tfidf_vect_sentences, columns=tfidf_feature_names)
df
```

```python
# Keras word encoding
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
```

```python
# 문장으로 부터 상위 100개 단어로 vocabulary 작성
tokenizer = Tokenizer(num_words=100, oov_token='<OOV>')
```

```python
# Word Index Vocabulary 작성
# sentences에 포함된 문장들을 기반을 단어의 토큰화를 수행하며, 각 단어에 고유한 인덱스를 할당
tokenizer.fit_on_texts(sentens)

# 각 단어에 부여된 고유 인덱스 추출
print(tokenizer.word_index)
print(tokenizer.index_word)
```

### tex의 sentence 변환 및 padding

*   texts_to_sequences : text list sodml 각 text를 수열 (sequence of integers)로 convert
    * 입력 : text(strings) list
    * 반환 : sequence list
*   pad_sequences : 동일한 길이로 sequence 를 zero padding



```python
# sentences 데이터를 시퀀스로 반환
sequences = tokenizer.texts_to_sequences(sentens)

# 시퀀스에 패딩 적용 (문장의 뒤쪽을 패딩하고, 필요시 뒤쪽을 잘라냄)
padded = pad_sequences(sequences, padding='post', truncating='post')

print(sequences)
print(padded)
```

### sequenced sentence를 word sentence로 환원

```python
# sequences 리스트에 있는 각 시퀀스를 처리
for sequence in sequences:
  sent = []
  for idx in sequence:
    sent.append(tokenizer.index_word[idx])
  print(' '.join(sent))
```

### One-Hot-Encoding 표현

```python
# 패딩된 스퀸스를 원-핫 인코딩으로 변환
to_categorical(padded)
```
